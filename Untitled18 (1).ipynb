```python
from crewai import Agent, Task, Crew, LLM
from crewai_tools import (
    FirecrawlSearchTool
)
from google.colab import userdata
SEARCH_API_KEY = userdata.get('FIRE_CRAWL_API_KEY')
GEMINI_API_KEY = userdata.get('GOOGLE_API_KEY') # Assuming you have stored your Gemini API key in Colab secrets under the name 'GOOGLE_API_KEY'

search_tool = FirecrawlSearchTool(api_key=SEARCH_API_KEY)

#Agent 1:Researcher
tech_researcher = Agent(
    role="Emerging Tech Researcher",
    goal="Identify top 5 emerging technologies gaining attention in 2025",
    backstory="A tech analyst who monitors innovation trends globally",
    tools=[search_tool],
    llm=LLM(model="gemini/gemini-2.0-flash", api_key=GEMINI_API_KEY)
)
#Agent 2: Summarizer
tech_summarizer = Agent(
    role="Tech Trend Summarizer",
    goal="Convert complex tech trends into short, digestible summaries",
    backstory="An expert in making technical topics easy to understand",
    tools=[],
    llm=LLM(model="gemini/gemini-2.0-flash", api_key=GEMINI_API_KEY)
)
task1 = Task(
    description="Search and list 5 trending emerging technologies in 2025 with brief details about each.",
    expected_output="Top 5 emerging technologies with description and potential impact.",
    agent=tech_researcher
)

task2 = Task(
    description="Summarize each of the 5 emerging technologies into one-liners for easy reading.",
    expected_output="List of 5 simplified, one-line tech summaries.",
    agent=tech_summarizer
)
tech_trends_crew = Crew(
    agents=[tech_researcher, tech_summarizer],
    tasks=[task1, task2],
    process="sequential"
)
print(tech_trends_crew.kickoff())
```
